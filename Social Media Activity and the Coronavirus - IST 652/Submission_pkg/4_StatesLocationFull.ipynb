```python
# Imported neccessary packages
import pandas as pd
```

```python
# Converted all_scores file to dataframe
df = pd.read_csv('all_scores.csv')
display(df)
```

```python
# Dealt with missing values
nan_value = float("NaN")
df.replace("", nan_value, inplace=True)
df
```

```python
# Sorted for locations that ended with state abbreviation
abv = df[df['location'].str.endswith((
          "AL", "AZ", "AR", "CA", "CO", "CT", "DE", "FL", "GA", "ID",
          "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
          "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
          "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", 
          "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"), na=False)]
abv
```

```python
# Returning only state abbreviations 
stateabv = []
for value in abv['location']:
    stateabv.append(value[-2:])
abv['location'] = stateabv
abv
```

```python
# Sorted for locations that ended with full state name and returned state
full = df[df['location'].str.endswith(("Alabama", "Arkansas", "Arizona", "California", "Colorado", "Connecticut", "Delaware", "Florida", "Georgia", "Iowa", "Idaho", "Illinois", "Indiana", "Kansas", "Kentucky", "Louisiana", "Massachusetts", "Maryland", "Maine", "Michigan", "Minnesota", "Missouri", "Mississippi", "Montana", "North Carolina", "North Dakota", "Nebraska", "New Hampshire", "New Jersey", "New Mexico", "Nevada", "New York", "Ohio", "Oklahoma", "Oregon", "Pennsylvania", "Rhode Island", "South Carolina", "South Dakota", "Tennessee", "Texas", "Utah", "Virginia", "Vermont", "Washington", "Wisconsin", "West Virginia", "Wyoming"), na=False)]
statefull = []
for value in full['location']:
    statefull.append(value.split(' ')[-1])
full['location'] = statefull
full = full.sort_values(by='location', ascending=True)
full
```

```python
# US state names and abbreviations dictionary
us_state_abbrev = {
    'Alabama': 'AL',
    'Arizona': 'AZ',
    'Arkansas': 'AR',
    'California': 'CA',
    'Colorado': 'CO',
    'Connecticut': 'CT',
    'Delaware': 'DE',
    'Florida': 'FL',
    'Georgia': 'GA',
    'Idaho': 'ID',
    'Illinois': 'IL',
    'Indiana': 'IN',
    'Iowa': 'IA',
    'Kansas': 'KS',
    'Kentucky': 'KY',
    'Louisiana': 'LA',
    'Maine': 'ME',
    'Maryland': 'MD',
    'Massachusetts': 'MA',
    'Michigan': 'MI',
    'Minnesota': 'MN',
    'Mississippi': 'MS',
    'Missouri': 'MO',
    'Montana': 'MT',
    'Nebraska': 'NE',
    'Nevada': 'NV',
    'New Hampshire': 'NH',
    'New Jersey': 'NJ',
    'New Mexico': 'NM',
    'New York': 'NY',
    'North Carolina': 'NC',
    'North Dakota': 'ND',
    'Ohio': 'OH',
    'Oklahoma': 'OK',
    'Oregon': 'OR',
    'Pennsylvania': 'PA',
    'Rhode Island': 'RI',
    'South Carolina': 'SC',
    'South Dakota': 'SD',
    'Tennessee': 'TN',
    'Texas': 'TX',
    'Utah': 'UT',
    'Vermont': 'VT',
    'Virginia': 'VA',
    'Washington': 'WA',
    'West Virginia': 'WV',
    'Wisconsin': 'WI',
    'Wyoming': 'WY'
}
```

```python
# Converting state full names to abbreviations 
for location in full['location']:
    for state in us_state_abbrev:
        if location.endswith(state):
            full['location'] = full['location'].replace([location],us_state_abbrev[state])
full
```

```python
# Compiling large dataframe
frames = [abv, full]
df_big = pd.concat(frames)
df_big
```

```python
# State abbreviations list
lis = ["AL", "AZ", "AR", "CA", "CO", "CT", "DE", "FL", "GA", "ID",
       "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
       "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
       "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", 
       "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]
```

```python
# Last check to see if all location values are state abbreviations
df_big = df_big[df_big['location'].isin(lis)]
df_big
```

```python
# Output dataframe to csv file
df_big.to_csv('all_states.csv')
```
