```python
# Importing neccessary packages
import pandas as pd
import numpy as np
```

```python
# Reading in csv file and converting to pandas dataframe object
csvFile = pd.read_csv('US_COVID_FINAL.csv') #read in covid dataset from the CDC 
states = ["AL","AZ", "AR", "CA", "CO", "CT", "DE", "FL", "GA", 
          "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
          "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
          "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", 
          "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"] #list of state Abbreviations we want to include
```

```python
# Returning dataframe with specific arguments 
df = pd.DataFrame(csvFile, #establish dataframe datatype, with columns matching below
columns=['submission_date', 'state', 'tot_cases','conf_cases','prob_cases','new_case','pnew_case','tot_death','conf_death','prob_death',\
         'new_death','pnew_death','created_at','consent_cases','consent_deaths'])
df['submission_date']= pd.to_datetime(df['submission_date'])#set datatype for dates
```

```python
# Converting time attribute to date time format
dfTime=df[df['submission_date']>'11/12/2020']
### define condition as cond_
cond=(dfTime["state"].isin(states))#set our condition to sort through the dataset, removing unwanted states and territories
dfTimeReduced=dfTime.loc[cond,:]#locate condition and include values that meet condition
```

```python
# Output dataframe to csv file
dfTimeReduced.to_csv('covidData.csv')
```
