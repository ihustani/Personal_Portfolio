```python
# Importing neccessary packages 
import pandas as pd
import folium
# Reading in neccessary files
covid = pd.read_csv('C:\\Users\\ian_u\\OneDrive\\Documents\\IST 652\\covidData.csv')
center = pd.read_csv('C:\\Users\\ian_u\\OneDrive\\Documents\\IST 652\\center.csv')
df = pd.read_csv('C:\\Users\\ian_u\\OneDrive\\Documents\\IST 652\\allStatesR.csv')
pops = pd.read_csv('C:\\Users\\ian_u\\OneDrive\\Documents\\IST 652\\2019Pops.csv')
covid
```

```python
# Returned pivoted dataframe, cases for each state by day
covidflip = covid.pivot(index="state", columns="submission_date",values="new_case")
covidflip 
```

```python
# Converted state names from index to column attribute, calculated change over time
covidflip['change'] = covidflip['2020-11-19'] - covidflip['2020-11-13']
covidflip['state'] = covidflip.index
covidflip.reset_index(drop=True, inplace=True)
covidflip
```

```python
# Created percent change of new cases map
data = covidflip.merge(center,  how='inner', left_on='state', right_on='Code')
change = 100 * (covidflip['change'] / covidflip['2020-11-13'])
data['change'] = change

CENTER_US = (39.8333333,-98.585522)
state_geojson = 'C:\\Users\\ian_u\\OneDrive\\Documents\\IST 652\\WMC3-us-states.json'
map = folium.Map(location=CENTER_US, zoom_start=4, tiles=' Open Street Map')

folium.Choropleth(geo_data=state_geojson, data=data, columns=['state','change'], 
                key_on='feature.id', fill_color='YlOrRd', legend_name='Percentage Change New Cases').add_to(map)

for row in data.to_records():
    pos = (row['Latitude'],row['Longitude'])
    message = f"Pct Change: {row['change']}"
    marker = folium.Marker(location=pos, 
                    popup=message
                          )
    map.add_child(marker)
map
```

```python
# Stripped time values of any ending
datetime = []
for value in df['time']:
    datetime.append(value.split(' ')[0])
df['time'] = datetime
df
```

```python
# Returned groupedby dataframe, twitter sentiment for each state by day
States = df.groupby(['location', 'time'], as_index=False)['neg', 'neu', 'pos', 'compound'].mean()
States = States.sort_values(by='pos', ascending=False)
States
```

```python
# Returned pivoted dataframe
sentflip = States.pivot(index="location", columns="time",values="compound")
sentflip
```

```python
# Converted state names from index to column attribute, calculated change over time
sentflip['change'] = sentflip['2020-11-19'] - sentflip['2020-11-13']
sentflip['location'] = sentflip.index
sentflip.reset_index(drop=True, inplace=True)
sentflip
```

```python
# Created percent change of sentiment map
data2 = sentflip.merge(center,  how='inner', left_on='location', right_on='Code')
change = 100 * (sentflip['change'] / sentflip['2020-11-13'])
data2['change'] = change
data2 = data2.dropna()


CENTER_US = (39.8333333,-98.585522)
state_geojson = 'C:\\Users\\ian_u\\OneDrive\\Documents\\IST 652\\WMC3-us-states.json'
map = folium.Map(location=CENTER_US, zoom_start=4, tiles=' Open Street Map')

folium.Choropleth(geo_data=state_geojson, data=data2, columns=['location','change'], 
                key_on='feature.id', fill_color='RdYlGn', legend_name='Percentage Change Compound Sentiment').add_to(map)

for row in data2.to_records():
    pos = (row['Latitude'],row['Longitude'])
    message = f"Pct Change: {row['change']}"
    marker = folium.Marker(location=pos, 
                    popup=message
                          )
    map.add_child(marker)
map
```

```python
# Returned groupedby dataframe, twitter sentiment for each state by hashtag
States2 = df.groupby(['location', 'bin'], as_index=False)['neg', 'neu', 'pos', 'compound'].mean()
States2 = States2.sort_values(by='pos', ascending=False)
States2
```

```python
# Returned groupedby dataframe with states and their average new_case count
avg = covid.groupby(['state'], as_index=False)['new_case'].mean()
avg
```

```python
# Created map of average new cases per capita
avg = avg.merge(center,  how='inner', left_on='state', right_on='Code')
avg = avg.merge(pops,  how='inner', left_on='State', right_on='NAME')
percap = avg['new_case'] / avg['POPESTIMATE2019']
avg['percap'] = percap

CENTER_US = (39.8333333,-98.585522)
state_geojson = 'C:\\Users\\ian_u\\OneDrive\\Documents\\IST 652\\WMC3-us-states.json'
map = folium.Map(location=CENTER_US, zoom_start=4, tiles=' Open Street Map')

folium.Choropleth(geo_data=state_geojson, data=avg, columns=['state','percap'], 
                key_on='feature.id', fill_color='YlOrRd', legend_name='Average New Cases Per Capita').add_to(map)

for row in avg.to_records():
    pos = (row['Latitude'],row['Longitude'])
    message = f"New Cases: {row['percap']}"
    marker = folium.Marker(location=pos, 
                    popup=message
                          )
    map.add_child(marker)
map
```
